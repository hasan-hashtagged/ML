```python
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.decomposition import PCA
```

```python
attr,label=[],[]
with open('dermatology.data') as fptr:
    reader=fptr.readlines()
    for line in reader:
        label.append(int(line.split(',')[-1]))
        attr_vector=line.split(',')[:-1]
        for i in range(len(attr_vector)):
            if attr_vector[i]=='?':
                attr_vector[i]=np.nan
            else:
                attr_vector[i]=int(attr_vector[i])
        attr.append(attr_vector)
X=np.array(attr)
Y=np.array(label)
imp=SimpleImputer(missing_values=np.nan,strategy='mean')
attr_clean=imp.fit_transform(X)
print(attr_clean)
print(Y)
```

```python
print(len(attr_clean))
```

```python
pca=PCA(n_components=10)
orthogonal_attr=pca.fit(attr_clean)
print(orthogonal_attr.components_)
```

```python
print(len(orthogonal_attr.components_[0]))
```
